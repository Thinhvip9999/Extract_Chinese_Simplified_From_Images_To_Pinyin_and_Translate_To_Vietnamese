```python
from google.colab import files
from PIL import Image
import pytesseract
import pinyin
from googletrans import Translator

# Hàm trích xuất chữ Trung Quốc từ ảnh (không xuống dòng)
def extract_chinese_text(image_path):
    img = Image.open(image_path)
    # Tắt xuống dòng và cấu hình xử lý các đoạn văn
    text = pytesseract.image_to_string(img, lang='chi_sim', config='--psm 6')
    # Loại bỏ các ký tự xuống dòng
    text = text.replace('\n', ' ')
    return text

# Hàm chuyển chữ Trung Quốc thành phiên âm
def get_pinyin(chinese_text):
    return pinyin.get(chinese_text, delimiter=' ')

# Hàm dịch sang tiếng Việt
def translate_to_vietnamese(text):
    translator = Translator()
    translated = translator.translate(text, src='zh-cn', dest='vi')
    return translated.text

# Chọn ảnh từ File Explorer
uploaded = files.upload()

# Lấy tên tệp đã tải lên
image_path = list(uploaded.keys())[0]

# Trích xuất chữ Trung Quốc từ ảnh
chinese_text = extract_chinese_text(image_path)
print(f"Chữ Trung Quốc trích xuất: {chinese_text}")

# Chuyển đổi chữ Trung Quốc thành phiên âm
pinyin_text = get_pinyin(chinese_text)
print(f"Phiên âm: {pinyin_text}")

# Dịch sang tiếng Việt
vietnamese_translation = translate_to_vietnamese(chinese_text)
print(f"Dịch sang tiếng Việt: {vietnamese_translation}")
```
